## PCA and ICA of Time Series Data

**Summary**: This notebook applys the pca and ica methods from section 2.2 in the report on the Nilearn time series data. The notebook visualizes and compares the different component networks regions between pca and ica.

```python
#Imports
import matplotlib.pyplot as plt
import numpy as np
from nilearn import datasets, plotting, image
from nilearn.maskers import NiftiMapsMasker
from sklearn.decomposition import FastICA
```

## I. PCA

First we create the same correlation matrix as stated in section 3.1 (reference pairwise correlations of time series data notebook)

```python
# Load the functional datasets
data = datasets.fetch_development_fmri(n_subjects=1)

# Load the Atlas
atlas = datasets.fetch_atlas_msdl()

atlas_filename = atlas["maps"] # Loading atlas image stored in 'maps'
labels = atlas["labels"] # Loading atlas data stored in 'labels'
```

```python
masker = NiftiMapsMasker(
    maps_img=atlas_filename,
    standardize="zscore_sample",
    standardize_confounds="zscore_sample",
    memory="nilearn_cache",
    verbose=5,
)
masker.fit(data.func[0])

mean_img = image.mean_img(data.func[0]) # The mean of the Time Series Data
time_series = masker.transform(data.func[0], confounds=data.confounds) # Time Series Data of Regions
```

```python
# Generate Correlation Matrix
correlation_matrix = time_series.T @ time_series / (len(time_series))
correlation_matrix.shape
```

```python
# To Preform PCA we take the eivectors of the correlation matrix
eival,eivec = np.linalg.eig(correlation_matrix)
```

```python
# First 3 components of our PCA
pca_first_comp, pca_second_comp, pca_third_comp = eivec[:, :3].T

print(pca_first_comp) #Example of first component
```

## II. ICA

### Step 1: Generate ICA components from time series

```python
# Using sklearn Fast ICA, we generate ICA components from time series

n = 3 # Number of components
seed = 42 # Seed for ICA

ica = FastICA(n_components = n, random_state = seed)
ica_components = ica.fit_transform(time_series.T).T
```

### Step 2: Normalizing and mapping back to 3D representation

```python
# Centering and Standarding ICA components (As mentioned in methods section 2.2)
ica_components -= ica_components.mean(axis=0)
ica_components /= ica_components.std(axis=0)
```

```python
ica_components.shape
```

```python
# First 3 components of our ICA
ica_first_comp, ica_second_comp, ica_third_comp = ica_components

print(ica_first_comp) #Example of first component
```

## III. Methods to visualize PCA and ICA components

Methods to parse the components and generate plots

```python
def get_component_regions(component_vector, threshold = 0.21):
    '''Gets the regions of the PCA/ICA network given the component cutoff value (Around 0.2)
    With the pca regions, we can visualize the network region generated by that component.

    Args:
        component_vector (list)  -- eigen vector of the pca component
        
    Keyword Args:
        threshold (float)        -- the cutoff that determines if the region is part of the network
    
    Returns:
    (set) each region in the pca component
    '''
    dic = dict(zip(labels, component_vector))
    
    return set([i for i in dic if abs(dic[i]) > threshold])
```

```python
def get_component_masker(component_vector, threshold = 0.21):
    '''Gets the masker of the component vector given the component cutoff value (Around 0.2).
    With the masker, we can visualize the pca component region scores in the network.

    Args:
        component_vector (list)  -- eigen vector of the pca component
        
    Keyword Args:
        threshold (float)        -- the cutoff that determines if the region is part of the network
    
    Returns:
    masker of each pca component
    '''
    
    masker = np.copy(component_vector)
    
    # Assigns component score to masker if above threshold, else assign 0
    vals = np.abs(component_vector) < threshold
    masker[vals] = 0
        
    return masker
```

```python
def plot_component(component_vector, title = "", style = "region", 
             display_mode = "ortho", cut_coords = None, threshold = 0.21):
    '''Plots the PCA/ICA component vector given the component cutoff value (Around 0.2) with the given style.

    Args:
        component_vector (list)  -- eigen vector of the pca component
        
    Keyword Args:
        title (str)              -- title of plot
        style (str)              -- style of the plot: "region" by regions, "score" by regions with score
        display_mode (str)       -- display_mode of the plot
        cut_coords (float)       -- the cut coords of the plot
        threshold (float)        -- the cutoff that determines if the region is part of the network
    
    Returns:
    plots the pca component
    '''

    fig, axs = plt.subplots(1, 1, figsize = (18, 10))

    if style == "region":
        
        # Get Regions of the component
        regions = get_component_regions(component_vector, threshold)

        # Builds the altas of the different regions
        atlas_regions = image.concat_imgs([image.index_img(atlas_filename, labels.index(i)) for i in regions])

        # Visualizes the regions of the component
        plotting.plot_prob_atlas(
            atlas_regions, draw_cross = False, title = title, axes=axs, 
            display_mode = display_mode, cut_coords = cut_coords
        )
    
    if style == "score":
        # Create Masker Images
        component_img = masker.inverse_transform(get_component_masker(component_vector))

        # Plotting PCA component with scores
        plotting.plot_stat_map(component_img, image.mean_img(data.func[0]), 
                               title = title, axes=axs, display_mode = display_mode, cut_coords = cut_coords
                              )
```

## IV. Visualizing PCA Components

```python
# Plot the regions of the first PCA component
plot_component(pca_first_comp, title = "PCA Component 1")
```

```python
# Plot the scores of the first PCA component network
plot_component(pca_first_comp, title = "PCA Component 1", style = "score")
```

```python
# Ploting regions of the first 2 PCA components
plot_component(pca_first_comp, title = "PCA Component 1", display_mode = "ortho", cut_coords = (0, 0, 0))
plot_component(pca_second_comp, title = "PCA Component 2", display_mode = "ortho", cut_coords = (0, 0, 0))
```

```python
# Ploting Moasics of the first 2 PCA components
plot_component(pca_first_comp, title = "PCA Component 1", display_mode = "mosaic")
plot_component(pca_second_comp, title = "PCA Component 2", display_mode = "mosaic")
```

## V. Visualizing ICA Components

```python
# Plot the regions of the first ICA component
plot_component(ica_first_comp, title = "ICA Component 1", threshold = 1.2)
```

```python
# Plot the scores of the first ICA component network
plot_component(ica_first_comp, title = "Component 1", style = "score", threshold = 1.2)
```

```python
# Ploting regions of the first 2 ICA components
plot_component(ica_first_comp, title = "ICA Component 1", display_mode = "ortho", 
               cut_coords = (0, 0, 0), threshold = 1.2)
plot_component(ica_second_comp, title = "ICA Component 2", display_mode = "ortho", 
               cut_coords = (0, 0, 0), threshold = 1.2)
```

```python
# Ploting Moasics of the first 2 ICA components
plot_component(ica_first_comp, title = "ICA Component 1", display_mode = "mosaic", threshold = 1.2)
plot_component(ica_second_comp, title = "ICA Component 2", display_mode = "mosaic", threshold = 1.2)
```

## VI. Compare ICA and PCA

```python
def plot_component_vs(component_vectors, titles, style = "region", 
             display_mode = "ortho", cut_coords = None, thresholds = [0.21, 1.2]):
    '''Plots the multiple PCA/ICA component vector given the component cutoff value (Around 0.2) with the given style.

    Args:
        component_vector (list)  -- eigen vector of the pca component
        
    Keyword Args:
        titles (str)              -- title of plot
        style (str)              -- style of the plot: "region" by regions, "score" by regions with score
        display_mode (str)       -- display_mode of the plot
        cut_coords (float)       -- the cut coords of the plot
        thresholds (float)        -- the cutoff that determines if the region is part of the network
    
    Returns:
    plots the pca component
    '''

    fig, axs = plt.subplots(1, 2, figsize = (18, 5))
    
    for i in range(len(component_vectors)):
        
        component_vector = component_vectors[i]

        if style == "region":

            # Get Regions of the component
            regions = get_component_regions(component_vector, threshold[i])

            # Builds the altas of the different regions
            atlas_regions = image.concat_imgs([image.index_img(atlas_filename, labels.index(i)) for i in regions])

            # Visualizes the regions of the component
            plotting.plot_prob_atlas(
                atlas_regions, draw_cross = False, title = titles[i], axes = axs[i], 
                display_mode = display_mode, cut_coords = cut_coords
            )

        if style == "score":
            # Create Masker Images
            component_img = masker.inverse_transform(get_component_masker(component_vector))

            # Plotting PCA component with scores
            plotting.plot_stat_map(component_img, image.mean_img(data.func[0]), 
                                   title = titles[i], axes = axs[i], display_mode = display_mode, cut_coords = cut_coords
                                  )
```

```python
# Plot pca and ica component 1 scores
plot_component_vs([pca_first_comp, ica_first_comp], titles = ["PCA Component 1", "ICA Component 1"], 
                  style = "score", cut_coords = (0, 0, 0))

# Plot pca and ica component 2 scores
plot_component_vs([pca_second_comp, ica_second_comp], titles = ["PCA Component 2", "ICA Component 2"], 
                  style = "score", cut_coords = (0, 0, 0))
```

```python
# Regions of the different components
print("PCA 1: " + str(get_component_regions(pca_first_comp)))
print("PCA 2: " + str(get_component_regions(pca_second_comp)))
print("ICA 1: " + str(get_component_regions(ica_first_comp, threshold = 1.2)))
print("ICA 2: " + str(get_component_regions(ica_second_comp, threshold = 1.2)))
```
