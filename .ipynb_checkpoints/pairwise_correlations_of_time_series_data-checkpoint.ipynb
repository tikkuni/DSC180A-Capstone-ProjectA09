## Generate Time Series Data from Nilearn

```python
#Imports
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from nilearn import datasets
from nilearn.maskers import NiftiMapsMasker
from nilearn.connectome import ConnectivityMeasure
from nilearn import plotting
```

```python
atlas = datasets.fetch_atlas_msdl()
# Loading atlas image stored in 'maps'
atlas_filename = atlas["maps"]
# Loading atlas data stored in 'labels'
labels = atlas["labels"]

# Load the functional datasets
data = datasets.fetch_development_fmri(n_subjects=1)
```

```python
masker = NiftiMapsMasker(
    maps_img=atlas_filename,
    standardize="zscore_sample",
    standardize_confounds="zscore_sample",
    memory="nilearn_cache",
    verbose=5,
)
masker.fit(data.func[0])
time_series = masker.transform(data.func[0], confounds=data.confounds)
```

## Pair wise correlation

In order to generate pairwise correlation in time series, we need to create a matrix between the time series data points. The formula for creating a correlation matrix is this:

Let A represent the time series matrix m x n with m being the number of regions and n the amount of data / time points

$$ C = \frac{A^T \cdot A}{n - 1}$$

This equation works as long as A is centered with its mean at zero and standardized with the standard deviation at 1. By using a standardized matrix we can ensure that the resulting non-diagonal elements fall between -1 and 1. Otherwise, the matrix generated would be a covariance matrix as mentioned above in performing PCA. Using the Nilearn package, we generated a correlation matrix from a resting state time series dataset.

```python
correlation_matrix = time_series.T @ time_series / (len(time_series) - 1)
correlation_matrix.shape
```

```python
np.fill_diagonal(correlation_matrix, 0)
plotting.plot_matrix(
    correlation_matrix, labels=labels, colorbar=True, vmax=0.8, vmin=-0.8, title = "Pairwise Correlation of Resting State")
```

This figure visualizes the correlation matrix with positive correlated values red, negatively correlated blue, and uncorrelated white. The resulting matrix is symmetric, and displays the correlations between each and every region.

```python
# Positive Correlation
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

l_aud_index = labels.index('L Aud')
r_aud_index = labels.index('R Aud')

axes[0].scatter(time_series[:,l_aud_index], time_series[:,r_aud_index])
axes[0].set(xlabel ='Left AUD bold activation', 
            ylabel = 'Right AUD bold activation', 
            title ='Left Auditory vs Right Auditory region')

axes[1].plot(time_series[:,l_aud_index], label = 'Left AUD', c = 'steelblue')
axes[1].plot(time_series[:,r_aud_index], label = 'Right AUD', c = 'indianred')
axes[1].set(xlabel ='Time', ylabel = 'Bold Activation', title ='Time Series')
axes[1].legend(loc="upper right")

print('Positive Correlation')
```

```python
# Negative Correlation
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

l_dmn_index = labels.index('L DMN')
r_ins_index = labels.index('R Ins')

axes[0].scatter(time_series[:,l_dmn_index], time_series[:,r_ins_index])
axes[0].set(xlabel ='Left DMN bold activation', 
            ylabel = 'Right INS bold activation', 
            title ='Left default network vs Right insular region')

axes[1].plot(time_series[:, l_dmn_index], label = 'Right INS', c = 'steelblue')
axes[1].plot(time_series[:, r_ins_index], label = 'Left DMN', c = 'firebrick')
axes[1].set(xlabel ='Time', ylabel = 'Bold Activation', title ='Time Series')
axes[1].legend(loc="upper right")

print('Negative Correlation')
```

```python
# No Correlation
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

motor_index = labels.index('Motor') # For Movement
broca_index = labels.index('Broca') # For Speech

axes[0].scatter(time_series[:,motor_index], time_series[:,broca_index])
axes[0].set(xlabel ='Motor bold activation', 
            ylabel = 'Broca bold activation', 
            title ='Motor vs Broca region')

axes[1].plot(time_series[:, motor_index], label = 'Motor', c = 'steelblue')
axes[1].plot(time_series[:, broca_index], label = 'Broca', c = 'darksalmon')
axes[1].set(xlabel ='Time', ylabel = 'Bold Activation', title ='Time Series')
axes[1].legend(loc="upper right")

print('No Correlation')
```
